```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import numpy as np
df=pd.read_csv("iris.csv")
df=pd.DataFrame(df)
x=df['sepal_length']
plt.hist(x,bins=20, color='green')
plt.title('sepal_length')
plt.xlabel('sepal_length')
plt.ylabel("count")
```
