```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("https://drive.google.com/u/0/uc?id=1H6wP97hHmgzppabdeTQrfRtDdghEfGls&export=download",encoding="ISO-8859-1")
print(df)
```

```python
#drop useless columns
df=df.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1)
```

```python
df=df.rename(columns={"v1":"label","v2":"sms"})
df.head()
```

```python
print(len(df))
```

```python
df.describe()
```

```python
#add column length of sms
df['lenghth']=df['sms'].apply(len)
```

```python
df.head()
```

```python
x=df['sms']
y=df['label']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.25,random_state=0)
```

```python
#CountVectorization Method
from sklearn.feature_extraction.text import CountVectorizer
count_vector=CountVectorizer()
```

```python
#fit training data
train_data=count_vector.fit_transform(x_train)
test_data=count_vector.transform(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
classifier=MultinomialNB()
classifier.fit(train_data,y_train)
```

```python
print(test_data)
```

```python
#prediction
y_pred=classifier.predict(test_data)
print(y_pred)
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
from wordcloud import WordCloud
#get spam messages
spam=df[df['label']=='spam']['sms'].str.cat(sep=",")
#get ham messages
ham=df[df['label']=='ham']['sms'].str.cat(sep=",")

wc=WordCloud(width=300,height=300,min_font_size=10,background_color="red")
#generate wordcloud for each type
spam_wc=wc.generate(spam)
#initialize wordcloud
#plot the wordcloud for spam
plt.figure(figsize=(5,5),facecolor=None)
plt.imshow(spam_wc)
plt.axis("off")
plt.title("common words in spam")
plt.tight_layout(pad=2)
plt.show()
#plot the wordcloud for ham
wc=WordCloud(width=300,height=300,min_font_size=10,background_color="orange")
spam_wc=wc.generate(ham)
plt.figure(figsize=(5,5),facecolor=None)
plt.imshow(spam_wc)
plt.axis("off")
plt.title("common words in spam")
plt.tight_layout(pad=0)
plt.show()

```
